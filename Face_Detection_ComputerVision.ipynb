```python
# Gray is special color which is that it is made by 1 color only that is combination of white and blacK(it is no color 0,0,0)
img1 = cv2.imread(r"C:\Users\sony\Desktop\pics\child.jpg")
img = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
plt.show()
```

### Spatial Filtering of Images

```python
# 1.Smoothing with Averaging Filter-Correlation/kernal'
plt.figure(figsize=(25,20))
img = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

plt.subplot(141)
plt.imshow(img)
plt.title('Original')

plt.subplot(142)
# Blur with a 5x5 averaging filter
blur1 = cv2.blur(img,(5,5))
plt.imshow(blur1)
plt.title('blured by 5x5')

plt.subplot(143)
# Blur with a 15x15 averaging filter
blur1 = cv2.blur(img,(15,15))
plt.imshow(blur1)
plt.title('blured by 15x15')

plt.subplot(144)
# Blur with a 25x25 averaging filter
blur1 = cv2.blur(img,(25,25))
plt.imshow(blur1)
plt.title('blured by 25x25')
```

### Thresholding for Binarization

```python
img1 = cv2.imread(r"C:\Users\sony\Desktop\pics\child.jpg")
img2 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
threshold,binary_img = cv2.threshold(img2,91,255,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize = (12,8))
# Display the output
plt.subplot(121)
plt.imshow(img2,cmap='gray')
plt.title('Original')

plt.subplot(122)
plt.imshow(binary_img,cmap='binary')
plt.title('Binary Image')
plt.show()
```

```python
# Create a black image
image =np.zeros((512,512,3),np.uint8)
plt.imshow(image),
plt.title('Blank Image')
plt.show()
```

```python
# cv2.line(image,starting point,end points,color code,width)
im2 = cv2.line(image,(0,200),(500,200),(255,105,180),10)
plt.imshow(im2)
plt.show()
```

```python
# cv2.line(image,starting point,end points,color code,width)
im2 = cv2.line(image,(0,200),(200,500),(255,105,180),10)
plt.imshow(im2)
plt.show()
```

```python
# Display the output

plt.figure(figsize = (12,8))
plt.subplot(121)
plt.imshow(img2,cmap='gray')
plt.title('Original')
```

```python
# Draw a rectangle (thickness )
img1 = cv2.imread(r"C:\Users\sony\Desktop\pics\child.jpg")

img2 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
cv2.rectangle(img2,
             (960,200),
             (1390,760),
             (255,255,34),
             thickness= 10,
             lineType = cv2.LINE_8
             )
plt.imshow(img2)
```

```python
# creating a circle
img1 = cv2.imread(r"C:\Users\sony\Desktop\pics\child.jpg")

img3 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
cv2.circle(img3, # image source
           (1170,490), # center
          260,         # radius
           (125,255,0), # color or border
          thickness=20,
          lineType=cv2.LINE_AA)

plt.imshow(img3)
plt.title('child-Circle')
```
